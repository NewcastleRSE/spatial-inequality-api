```python
from spineq.optimise import optimise, calc_oa_weights
from spineq.plotting import plot_oa_weights, plot_optimisation_result

import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
n_sensors = 20
theta = 500

# colour scale range
vmin = 0
vmax = 0.01
```

## Total Population Only

```python
age_weights = 1
population_weight = 1
workplace_weight = 0

weights = calc_oa_weights(age_weights=age_weights,
                          population_weight=population_weight,
                          workplace_weight=workplace_weight)

plot_oa_weights(weights, title="Weights: OA Total Residential Population",
                vmin=vmin, vmax=vmax)
```

```python
result = optimise(n_sensors=n_sensors,
                  theta=theta,
                  age_weights=age_weights,
                  population_weight=population_weight,
                  workplace_weight=workplace_weight)

plot_optimisation_result(result)
```
